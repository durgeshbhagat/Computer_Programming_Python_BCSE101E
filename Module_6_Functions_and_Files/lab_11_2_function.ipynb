#### Prime number code using function

```python
### without using flag variable
def check_prime(n):
  remainder_list = []
  condition_list = []
  for i in range(2,n):
    print(i)
    remainder_list.append((i, n%i))
    condition_list.append( n%i)
    if n% i ==0:
        break
  print('Condition list:' , condition_list)
  print('remainder_list :' , remainder_list)
  if 0 in condition_list:
    print('Not prime')
  else:
    print('Prime')
```

```python
check_prime(9)
```

```python
check_prime(11)
```

```python
check_prime(7)
```

```python
# using flag variable
def check_prime_2(n):
  is_prime = True # flag
  for i in range(2,n):
    print(i)
    if n% i ==0:
      is_prime = False
      break
  if is_prime:
    print('Prime')
  else:
    print('Not Prime')
```

```python
check_prime_2(8)
```

```python
check_prime_2(7)
```

```python
check_prime_2(87)
```

#### Factorial

```python
# recursive factorial
def recursive_fact(n):
  """ calculate factorial of an positive integer using recusrion
  """
  # base case
  if n == 1 or n==0:
    return 1
  else:
    # recursive case
    return n* recursive_fact(n-1)
```

```python
print(recursive_fact(1))
```

```python
print(recursive_fact(0))
```

```python
print(recursive_fact(5))
```

```python
# iterative factorial
def iterative_fact(n):
  """
  factorial of an positive integer
  """
  product =1 
  for num in range(2, n+1):
    product = product * num
  return product
```

```python
print(iterative_fact(0))
```

### Fibbonacci
- to find  $n^{th}$ number from fibbonacci series
- Fibbonacci series: 0, 1, 1, 2, 3, 5, 8, ...
- $F_n = F_{n-1} + F_{n-2}$

```python
def fib_recursive(n):
  if n==0: # base case
    return 0 
  elif n==1: # base case
    return 1
  else: # recursive case
    return fib_recursive(n-1) + fib_recursive(n-2)
```

```python
print(fib_recursive(5))
```

#### Find the GCD of two number using recursive and iterative approach

```python
def gcd(num1, num2):
  """
  Euclidian algorithm using substraction
  """
  if num1 == num2:
    return num1
  elif num1 > num2:
    return gcd(num1-num2, num2)
  else:
    return gcd(num1, num2 - num1)
```

```python
print(gcd(16,24))
```

```python
num1 = int(input('Enter a number: '))
num2 = int(input('Enter a number: '))
num_gcd  = gcd(num1, num2)
print(f'num1= {num1}, num2 = {num2}; gcd={num_gcd}')
```

```python
def gcd2(num1, num2):
  # Modified Euclidian algorithm using divison
  if num2 == 0:
    return num1
  else:
    return gcd(num2, num1 % num2)
```

```python
num_gcd = gcd2(35,45)
print(num_gcd)
```

#### Find LCM of two numbers

```python
def lcm(num1, num2):
  num_gcd = gcd(num1, num2)
  num_lcm = (num1*num2) / num_gcd
  return num_lcm
  
```

```python
num_lcm = lcm(12,18)
print(num_lcm)
```
