#### To check whether a number is *Even* or *Odd*

PART 1: Function Definition
- It tells us *How to do a particular task* ?

```python
# basic function - Function defintion
def check_even(num):
  if num%2==0:
    print('Even number')
  else:
    print('ODD number')
```

PART 2: Function Call
- It *call* or *performs* a particular task

```python
# Function call
check_even(5)
```

```python
check_even(8)
```

```python
# is this a function definition or function call
def Funpython():
  print('Python is fun ❤️')
```

```python
Funpython()
```

#### To check whether a number is *prime* or not

```python
def check_prime(n):
  is_prime = True
  for i in range(2,n):
    if n%i == 0: 
      is_prime = False
      break
  return is_prime  
```

```python
print(check_prime(8))
```

```python
print(check_prime(2))
```

```python
print(check_prime(11))
```

```python
print(check_prime(49))
```
