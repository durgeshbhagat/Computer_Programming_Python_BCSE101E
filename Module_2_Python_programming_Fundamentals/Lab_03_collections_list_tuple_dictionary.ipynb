# Collections :- list, tuple, dictionary, set

## List
- List is collection of heterogeneous items
- List size can be changed dynamically

```python
# declaration of a list
my_list = [1,7,58,24, 'BCSE101E', 'NLP', 99.8,'Windows']
```

```python
# accessing elements of list - list index starts from 0
print(my_list[2])
print(my_list[1])
print(my_list[5])
```

```python
# re-assigning an list index
print('before re-assigning', my_list)
my_list[3] = 'new_item'
print(' After re-assigning' , my_list)
```

```python
# slicing and striding over list
print(my_list[2:7])
print(my_list[2:7:2])
```

```python
# List can also be accessed with negative indxing
print(my_list[-1])
print(my_list[-2])
print(my_list[-3])
print(my_list[-5])
```

```python
my_list.append(199)
print(my_list[-1])
```

```python
print(my_list[:4])
my_list.insert(2, 96)
print(my_list[:4])
```

## Tuples
- tuple is also a collection of heteregenous elements

```python
#tuple declaration
my_tup = (1,'ram', 67, 3.14, 'BCSE101E')
print(my_tup)
```

```python
# Accessing elements of tuple - tuple indexing start from zero (0)
print(my_tup[1])
print(my_tup[4])
print(my_tup[0])
```

```python
list_loc = [ (2,3) , (4,5), (7,13)]
print(list_loc[1][0])
```
