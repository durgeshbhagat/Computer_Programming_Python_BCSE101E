## Calculate simple interst, given principal amount (p), rate of interest(r) and time in years respectively. Consider p and t as intergers and r as float

```python
p = int(input()) # converting p from string to integer
r = float(input()) # converting from string to float
t = int(input()) # converting from string to integer
si = (p*r*t) /100
print(f'simple interest is : {si}')
```
