**Calculate the total amount returned on the**
- principal amount (P),
- rate of interest (r),
- and time (in year)

**assuming the rates are compounded annually.**


![compund interest formulae](https://drive.google.com/uc?export=view&id=1r6a_6EJSgRRn9C2mhNR7tAk3vFpn7KAM)

```python
p = int(input())
r  =int(input())
t = int(input())
n = 1 # time interveal after which interest is calculated and added to principal amount e.g compunded annually=> n=1
AR = p* (1 + (r/ (100*n)))**(n*t)
print(f"Amount returned : {AR}")
```

### **Q2: Given the height of a person in feet and inches, convert it into cms**
- **where `1 feet = 12 inch` and `1 inch= 2.54 cm`**

```python
h_ft = int(input()) # hright of a person in feet
h_in = int(input()) # height of a erson in inch

h_in_total = 12*h_ft + h_in
h_cm = 2.54 * h_in_total
print(f'height in cm = {h_cm}')
```

### **Q4. Given a quadratic equation in the form of `a *x^2 + b*x + c`***.
**Find the roots of this quadratic equation where a,b, and c are taken from user as integer input**.

```python
import math
a = int(input())
b = int(input())
c = int(input())
d_square =  b*b - 4*a*c
print(f'd sqayre = {d_square}')
if d_square ==0:
  root = -b /( 2*a)
  print('Equal roots {root}')
elif d_square >0:
  root_1 = (-b + math.sqrt(d_square) ) / (2*a)
  root_2 = (-b - math.sqrt(d_square)) / (2*a)
  print("Root are real and different")
  print(root_1)
  print(root_2)
else:
  real_part = (-b) / (2*a)
  img_part = math.sqrt(-1 * d_square) / (2*a)

  root_1 = complex(real_part , img_part)
  root_2 = complex(real_part, -1*img_part)
  print('Roots are imaginary')
  print(root_1)
  print(root_2)
```
