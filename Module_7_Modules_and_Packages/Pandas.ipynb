```python
st = """22BCE3292	SHREYAS SANTOSH KUMAR	shreyas.santosh2022@vitstudent.ac.in 
22BCE3478	KURIAN AVARAN JOJO	kurianavaran.jojo2022@vitstudent.ac.in
22BCE2612	VARIN GURU	varin.guru2022@vitstudent.ac.in
22BCE2521	AADHAVAN PONNIAH ANBUMUNEE	aadhavan.ponniah2022@vitstudent.ac.in
22BCE2652	TANISHQ RAVISHANKAR TIWARI	tanishq.ravishankar2022@vitstudent.ac.in
22BCE3184	SAUMYA SAANVI	saumya.saanvi2022@vitstudent.ac.in
22BCE2893	SAATVIK TIWARI	saatvik.tiwari2022@vitstudent.ac.in
22BCE2842	RISHI KEDAR BONDRE	rishikedar.bondre2022@vitstudent.ac.in
22BCE2941	ATULYA JAYANT SAHANE	atulyajayant.sahane2022@vitstudent.ac.in
22BCE3064	ARYAN BHIRUD	aryan.bhirud2022@vitstudent.ac.in
22BCE0166	CHARAN ADITYA RAVICHANDRAN	charan.aditya2022@vitstudent.ac.in
22BCE3034	M RISHI KRISHNAN	rishikrishnan.m2022@vitstudent.ac.in"""
```

```python
student_list = []
line_list = st.split('\n')
for i, line in enumerate(line_list):
  cur_line_list = line.strip().split()
  r_no = cur_line_list[0]
  name = ' '.join(cur_line_list[1:-1])
  email = cur_line_list[-1]
  student_list.append([ r_no, name, email])
```

```python
for i, student in enumerate(student_list):
  print(i, student)
```

```python
df = pd.DataFrame(student_list, columns= ['RNo', 'Name', 'Email'])
```
