- For loop syntax

- Nested for loop
    - dependent nested for loop
    - independent nested for loop

- Examples

for loop syntax over sequqnce/iterable/collections

```python
lang_list = ['Python', 'C',  'C++', 'Java', ' Maths']
for item in lang_list:
  print(item)
```

```python
name = "Durgesh Kumar"
word_list = name.split(' ')
print(word_list)
print(len(word_list))
```

```python
name = "Durgesh Kumar"
space_count =0
for ch in name:
  if ch == ' ':
    space_count += 1
print(f'space_count={space_count}')
print(f'word_count={space_count+1}')
```

### `for` with item and index using `enumerate`

```python
lang_list = ['Python', 'C',  'C++', 'Java', ' Maths']
for index, item in enumerate(lang_list):
  print(index, item)
```

```python
lang_list = ['Python', 'C',  'C++', 'Java', ' Maths']
for index, item in enumerate(lang_list, 1):
  print(index, item)
```

```python
lang_list = ['Python', 'C',  'C++', 'Java', ' Maths']
for index, item in enumerate(lang_list[2:], 1):
  print(index, item)
```

## prime number

```python
num = int(input())
for num == int():
  if num%1==0 and num%num ==0:
    print('IS PRIME')
  else:
    print('NOT PRIME')
```

```python
num = int(input()) #1
if num >1:
  for i in range(2, int(num/2) +1 ):
    if num % i  ==0: # num%1 ==0
      print('IS PRIME')
    else:
      print('NOT PRIME')
```

```python
num = int(input()) #1
if num >1:
  for i in range(2, int(num/2) +1 ):
    if num % i  ==0: # num%1 ==0
      print('NOT PRIME')
      break
  else:
      print('IS PRIME')
```

```python
num = int(input())
is_prime = True
for i in range(2, num):
  if num %i ==0:
    is_prime = False
    break

if is_prime:
  print(f'{num} is prime')
else:
  print(f'{num} is NOT prime')
```

## Nested for loop
- independent loop
- dependent loop

```python
# independent loop
list1 = ['I am ', 'You are ']
list2 = ['healthy', 'fine', 'geek']
for item1 in list1:
  for item2 in list2:
    print(item1, item2)
```

```python
# independent loop
for i in range(4):
  for j in range(100, 102):
    print(i, j)
```
