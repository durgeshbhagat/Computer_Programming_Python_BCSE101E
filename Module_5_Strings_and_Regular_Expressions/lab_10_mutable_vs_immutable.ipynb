# Mutable vs immutable
## `Objects` of built-in type that are mutable are:

- Lists
- Sets
- Dictionaries
- User-Defined Classes (It purely depends upon the user to define the characteristics) 

## `objects` of built-in type that are immutable are:

- Numbers (Integer, Rational, Float,Decimal, Complex & Booleans)
- Strings
- Tuples
- Frozen Sets
- User-Defined Classes (It purely depends upon the user to define the characteristics)

## Objects in Python
- In Python, everything is treated as an object. 
- Every object has these three attributes:
  i) `Identity` – This refers to the address that the object refers to in the computer’s memory.

  ii) `Type` – This refers to the kind of object that is created. For example- integer, list, string etc. 

iii) ``Value`` – This refers to the value stored by the object. 

For example – 
myList=[1,2,3] would hold the numbers 1,2 and 3

**Note**: While ID and Type cannot be changed once it’s created, values can be changed for Mutable objects.

### Mutable objects in python

```python
cities = [' Delhi', 'Kolkata', 'Chennai', 'Jaipur']
for city in cities:
		print(city, end=', ')
```

```python
print(id(cities))
```

```python
print(hex(id(cities)))
```

```python
print(hex(id(cities)))
cities.append('Surat')
for city in cities:
	print(city, end=', ')
print('')
print(hex(id(cities)))
```

### Immutable Objects in Python

```python
days = ('Sunday', 'Monday', 'Tuesday',  \
        'Wednesday', 'Thursday', \
        'Friday', 'Saturday')
```

```python
for d in days:
  print(d)
```

```python
print(hex(id(d)))
print(hex(id(days)))
```

```python
days[-1] = 'Deepawali'
```

```python
days += 'Deepawali'
```
