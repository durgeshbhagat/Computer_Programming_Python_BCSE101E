## Q1. WAF to count the occurrence  of every character  in a string.
   a. Find the most frequent character
   
   b. Find the least frequent character.


```python
def count_char_freq(st):
  char_dict= {}
  for ch in st:
    if ch not in char_dict:
      char_dict[ch] = 0
    char_dict[ch] += 1
  return char_dict
```

```python
my_st = 'I am learning python at SJT 515 Lab'
char_dict_new = count_char_freq(my_st)
print(my_st)
print(char_dict_new)
```

```python
max(char_dict_new.items(), key=lambda x: x[1] )[0]
```

```python
sorted(char_dict_new.items(), key=lambda x: x[1], reverse=True)[0]
```

```python
min(char_dict_new.items(), key=lambda x: x[1] )
```

## Find the number of words in a string considering SPACE as word separator.

```python
def count_number_of_words(st):
  word_list = st.strip().split(' ')
  print(word_list)
  return word_list
```

```python
my_st = '  \n I am learning python at SJT 515 Lab \t'
words_list = count_number_of_words(my_st)
print(words_list)
print(len(words_list))
```

## String Formatting

#### left, center, and right alignment

```python
String="Python"
print("Left,centre and right alignment by formatting:")
print("|{:<15}|{:^15}|{:>15}|".format(String,'is','fun'))
```

#### decimal to binary, floating point representation

```python
print("Binary representation of 16 is: {0:b}".format(10))
print("Exponent form of of 13.56 is: {0:e}".format(13.56))
print("Rounding of 1/3 to 3 decimal places: {0:.3f}".format(1/3))
```
